# **Categorical** Analysis

Gender-wise

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.2f', figsize = (6,6))
# The plot visually represents the proportion of males and females in the dataset

# '.size()' calculates the no. of occurrences for each unique value in 'Sex' group
```

Chest Pain Type-wise

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.2f', figsize = (6,6))

# This plot visually represents the proportion of each chest pain type:
# Asymptomatic(ASY), Atypical Angina(ATA), Non-Anginal Pain(NAP) & Typical Angina(TA) in the dataset
```

RestingECG type-wise

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.2f', figsize = (6,6))

# This plot visually represents the proportion of each resting Electrocardiogram type:
# Left Ventricular Hypertrophy(LVH), Normal ECG, ST-T wave abnormality(ST) in the dataset
```

ST Slope type-wise

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
# The plot visually represents the proportion of each ST slope type:
# Flat, Up & Down in the dataset which describes the shape and direction of ST segment
```

'HeartDisease' distribution

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
# The plot represents the proportion of individuals with(1) & without heart disease(0) in the dataset
```

# **Violin** Plots using **Seaborn** Library

A violin plot is a method of plotting numeric data, similar to box plot.

It shows the probability density of the data at different values, smoothed by a kernel density estimator.

Age Distribution

```python
sns.violinplot(x = df['Age'])
# The plot helps to visualize the distribution of ages & see where the data is concentrated.
```

Gender-wise 'HeartDisease' distribution

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])

# The plot shows the distribution of 'HeartDisease' for each gender.
# It allows us to compare the distribution of heart disease between males & females in the dataset
```

Age-wise 'HeartDisease' distribution

```python
sns.violinplot(y = df['Age'], x = df['HeartDisease'])

# The plot shows the distribution of 'Age' for individuals with(1) and without(0) heart disease
# It allows us to compare the age distribution between these two groups in the dataset.
```

'Cholesterol' level-wise 'HeartDisease' distribution

# **Heatmap** using **Seaborn** Library

A **heatmap** is a graphical representation of data where the individual *values* contained in a matrix are represented as *colors*.

 In this case, the *colors* represent the **correlation coefficients** between the pairs of columns. *'Darker colors'*  typically indicate **weaker** correlations, while *'lighter colors'*  indicate **stronger** correlations.

 The *color bar* on the right shows the **mapping** between colors and correlation values.

```python
df[['Age', 'RestingBP','Cholesterol', 'FastingBS','MaxHR','Oldpeak','HeartDisease']].corr()

# Here, we first select a subset of the DataFrame containing the columns:
#'Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', and 'HeartDisease'

# Then, the pairwise correlation between these columns is calculated using the '.corr()' method
# The result is a correlation matrix
```

```python
sns.heatmap(df[['Age', 'RestingBP','Cholesterol', 'FastingBS','MaxHR','Oldpeak','HeartDisease']].corr())
# The heatmap of the correlation matrix calculated in the previous step
```

# **Pair-plots** using **Seaborn** Library

**Pairplots** are used to create a *grid of pairwise relationships*  in the dataframe.

The **pairplot** is a way to quickly visualize the distributions of **individual** variables and the *'relationships'* between **pairs of variables** in the dataset.

Generally, on the *diagonal* of the grid, a **histogram** or **KDE plot** is shown for each individual column, displaying its distribution.
In the *off-diagonal* plots, **scatter plots** are shown for each pair of columns, illustrating the relationship between them.

'Age - HeartDisease' Pairplot

```python
# Pair plot to show pairwise relationship between 'Age' & 'HeartDisease' columns
sns.pairplot(df[['Age', 'HeartDisease']])
```

'Age - RestingBP - HeartDisease' Pairplot

```python
sns.pairplot(df[['Age', 'RestingBP', 'HeartDisease']], hue='HeartDisease')

# This plot helps to understand how the distributions and relationships between 'Age' & 'RestingBP'
# differ for individuals with and without heart disease

# hue='HeartDisease' is used to color the points in the scatter plots & the density curves
# in the diagonal plots based on the value in the 'HeartDisease' column (0 or 1).
#This allows us to visually compare the distributions & relationships for individuals with & without heart disease
```

'Age - Cholestrol - FastingBS - HeartDisease' Pairplot

```python
sns.pairplot(df[['Age', 'Cholesterol', 'FastingBS', 'HeartDisease']], hue='HeartDisease')

# The plot allows us to explore the pairwise relationships and individual distributions of:
# 'Age', 'Cholesterol', and 'FastingBS', while also seeing that how these patterns
# differ based on whether an individual has heart disease
```

'Age - MaxHR - Oldpeak - HeartDisease' Pairplot

```python
sns.pairplot(df[['Age', 'MaxHR', 'Oldpeak', 'HeartDisease']])

# The plots are useful to examine the relationships between  the 'Age',
# 'MaxHR', 'Oldpeak', and 'HeartDisease' columns
```

Pairplot using all numerical Columns

```python
sns.pairplot(df)
# Here, a matrix of plots is generated where each numerical column in the DataFrame is
# paired with every other numerical column
```

# **Jointplots** using **Seaborn** Library

**Jointplots** are used for exploring the *'bivariate'* relationship (how two variables relate to each other) while also showing the *'univariate'* distributions (the distribution of each variable separately) on the margins.

'Age - MaxHR' Jointplot

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')

# The plot helps to visualize the relationship between Age & Maximum Heart Rate,
# and see where the data is most concentrated.

# 'hex'is used to create a hexagonal binning plot which is useful for visualizing
# the density of points in areas where they overlap;
# the darker the hexagon, the more data points fall within that bin
```

'Age - Cholestrol' Jointplot

```python
sns.jointplot(x = 'Age', y = 'Cholesterol', data = df, kind = 'scatter')

# The plot helps to visualize the relationship between Age and Cholesterol levels
# 'scatter' specifies scatter plot, showing individual data points for each pair of 'Age' & 'Cholesterol' values
```

'RestingBP - MaxHR' Jointplot

```python
sns.jointplot(x = 'RestingBP', y = 'MaxHR', data = df, kind = 'hex')

#The plot helps to visualize the relationship between Resting Blood Pressure and Maximum Heart Rate,
# and see where the data is most concentrated.
```
